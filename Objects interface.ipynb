## seaborn.objects 인터페이스

* 네임 seaborn.objects스페이스는 Seaborn 플롯을 만들기 위한 완전히 새로운 인터페이스로 버전 0.12에서 도입되었습니다. 데이터 변환 및 플로팅을 위한 구성 가능한 클래스 모음으로 구성된 보다 일관되고 유연한 API를 제공합니다. 기존 기능과 달리 seaborn새 인터페이스는 matplotlib로 드롭다운하지 않고 종단 간 플롯 사양 및 사용자 지정을 지원하는 것을 목표로 합니다(필요한 경우 계속 가능함).

* 개체 인터페이스는 현재 실험적이며 불완전합니다. 진지하게 사용하기에 충분히 안정적이지만 확실히 일부 거친 가장자리와 누락된 기능이 있습니다.

## 플롯 지정 및 데이터 매핑

* 개체 인터페이스는 다음 규칙에 따라 가져와야 합니다.

```python
import seaborn as sns
import seaborn.objects as so
```

* 네임 seaborn.objects스페이스는 모든 관련 클래스에 대한 액세스를 제공합니다. 가장 중요한 것은 Plot. 개체를 인스턴스화하고 해당 메서드를 호출하여 플롯을 지정합니다 Plot. 간단한 예를 보자:

```python
penguins = sns.load_dataset('penguins')
```

```python
(
    so.Plot(penguins, x="bill_length_mm", y="bill_depth_mm")
    .add(so.Dot())
)
```

* 산점도를 생성하는 이 코드는 꽤 친숙하게 보일 것입니다. 를 사용할 때와 마찬가지로 seaborn.scatterplot()깔끔한 데이터 프레임( )을 전달 penguins하고 해당 열 중 두 개를 플롯의 x및 좌표에 할당했습니다. y그러나 차트 유형으로 시작한 다음 일부 데이터 할당을 추가하는 대신 여기서는 데이터 할당으로 시작한 다음 그래픽 요소를 추가했습니다.

## 속성 설정 

* 클래스 는 데이터 값을 그래픽으로 나타내는 개체 Dot의 예입니다 . Mark각 마크에는 모양을 변경하기 위해 설정할 수 있는 여러 속성이 있습니다.

```python
(
    so.Plot(penguins, x="bill_length_mm", y="bill_depth_mm")
    .add(so.Dot(color="g", pointsize=4))
)
```

## 매핑 속성 

* seaborn의 기능과 마찬가지로 데이터 값을 다양한 그래픽 속성에 매핑하는 것도 가능합니다.

```python
(
    so.Plot(
        penguins, x="bill_length_mm", y="bill_depth_mm",
        color="species", pointsize="body_mass_g",
    )
    .add(so.Dot())
)
```

* hue이 기본 기능이 새롭지는 않지만 함수 API와의 중요한 차이점은 속성을 직접 설정하는 동일한 매개 변수 이름을 사용하여 속성이 매핑된다는 것입니다( vs. 등 을 사용하는 대신 color). 중요한 것은 속성이 정의된 위치Dot 입니다. 초기화할 때 값을 전달하면 직접 설정되지만 설정 시 변수를 할당하면 해당 데이터가 매핑Plot 됩니다 .

* 이러한 차이점 외에도 개체 인터페이스를 사용하면 훨씬 더 광범위한 마크 속성을 매핑할 수 있습니다.

```python
(
    so.Plot(
        penguins, x="bill_length_mm", y="bill_depth_mm",
        edgecolor="sex", edgewidth="body_mass_g",
    )
    .add(so.Dot(color=".8"))
)
```
